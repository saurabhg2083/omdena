```python
# https://www.buyrentkenya.com/houses-for-rent

import requests
from bs4 import BeautifulSoup as bs
from urllib.request import Request, urlopen
```

```python
#houses_for_rent_url = "https://www.buyrentkenya.com/houses-for-rent?page=2"
```

```python

url="https://www.buyrentkenya.com/houses-for-rent"
req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})

web_byte = urlopen(req).read()

webpage = web_byte.decode('utf-8')

html = bs(webpage, "html.parser")
html
```

```python
bigboxes = html.findAll("div", {"class": "mb-5 mx-3 md:mx-0"})
bigboxes
```

```python
# one page has maximum 0 to 16 listing
box = bigboxes[0]
box
```

```python
detailLink = "https://www.buyrentkenya.com" + box.div.div.div.a['href']

#flex flex-col justify-between md:w-2/3 px-5 py-4
detailLink
```

```python
req = Request(detailLink, headers={'User-Agent': 'Mozilla/5.0'})

web_byte = urlopen(req).read()

webpage = web_byte.decode('utf-8')

html = bs(webpage, "html.parser")
html
```
